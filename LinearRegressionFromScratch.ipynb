```python
import numpy as np
```

```python
data = np.loadtxt('data.csv',delimiter=",")
```

```python
x = data[:,0]
y = data[:,1]
```

```python
x.shape
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(x,y,test_size = 0.3)
X_train.shape
```

```python
def fit(x,y):
  num = (x*y).mean() - x.mean()*y.mean()
  denim = (x**2).mean() - x.mean()**2
  m = num/denim
  c = y.mean() - (m*x).mean()
  return (m,c)
```

```python
def predict(x,m,c):
  return m*x + c
```

```python
def score(y_true,y_pred):
  u = ((y_true-y_pred)**2).sum()
  v = ((y_true-y_true.mean())**2).sum()
  result = 1-(u/v)
  return result
```

```python
def cost(x,y,m,c):
  return ((y-m*x-c)**2).mean()
```

```python
m,c = fit(X_train,Y_train)
#test data
y_test_pred = predict(X_test,m,c)
print("Test Score:",score(Y_test,y_test_pred))

#train data
y_train_pred = predict(X_train,m,c)
print("Train Score:",score(Y_train,y_train_pred))
print("Cost on training data",cost(X_train,Y_train,m,c))
print("M, C",m,c)
```

```python
# INBUILT LINEAR REGRESSION
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg = LinearRegression()
```

```python
#Reshape the 1D array data to 2D array bcz fit require 2d array for LinearRegression
x_train = X_train.reshape(-1,1)
x_test = X_test.reshape(-1,1)
y_train = Y_train.reshape(-1,1)
y_test = Y_test.reshape(-1,1)
```

```python
alg.fit(x_train,y_train)
m = alg.coef_[0]
c = alg.intercept_
#test data
y_te_pred = alg.predict(x_test)
print("Test Score by Inbuilt Model is:",alg.score(y_te_pred,y_test))

#train data
y_ta_pred = alg.predict(x_train)
print("Train Score by Inbuilt Model is:",alg.score(y_ta_pred,y_train))
print("M, C",m,c)
```
